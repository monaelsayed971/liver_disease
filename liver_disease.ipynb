```python
from sklearn.utils import resample
from category_encoders import BinaryEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import RandomOverSampler

# Upsample minority class
df_clean_minority_upsampled = resample(df_clean_minority,
                                       replace=True,
                                       n_samples=416,
                                       random_state=42)

# Combine majority class with upsampled minority class
df_clean_upsampled = pd.concat([df_clean_majority, df_clean_minority_upsampled])

# Separate features (X) and target variable (y)
X_train = df_clean_upsampled.drop('Dataset', axis=1)
y_train = df_clean_upsampled['Dataset']

# Define preprocessing for numerical data
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

# Define preprocessing for categorical data
categorical_transformer = Pipeline(steps=[('binary', BinaryEncoder())])

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, ['Age', 'Total_Bilirubin', 'Direct_Bilirubin', 'Alkaline_Phosphotase',
                                        'Alamine_Aminotransferase', 'Aspartate_Aminotransferase', 'Total_Protiens',
                                        'Albumin', 'Albumin_and_Globulin_Ratio']),
        ('cat', categorical_transformer, ['Gender'])])

# Define the resampling step
resampler = RandomOverSampler(random_state=42)

# Create an imbalanced-learn pipeline with resampling
pipeline = ImbPipeline(steps=[('preprocessor', preprocessor),
                              ('resampler', resampler)])

# Fit the pipeline on the training data
X_train_preprocessed, y_train_preprocessed = pipeline.fit_resample(X_train, y_train)

# Preprocess the test data
X_test_preprocessed = pipeline.named_steps['preprocessor'].transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X_train_preprocessed)
explained_variances = pca.explained_variance_ratio_
cumulative_variances = np.cumsum(explained_variances)
threshold = 0.95
num_components = np.argmax(cumulative_variances >= threshold) + 1
pca = PCA(n_components=num_components)
X_train_pca = pca.fit_transform(X_train_preprocessed)
X_test_pca = pca.transform(X_test_preprocessed)
```

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
accuracy_test = model.score(X_test, y_test)
print(f"Model accuracy for test : {accuracy_test}")
accuracy_train = model.score(X_train, y_train)
print(f"Model accuracy for train : {accuracy_train}")
print(classification_report(y_test, y_pred))
```
